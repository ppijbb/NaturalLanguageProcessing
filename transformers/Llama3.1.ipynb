````python
import torch
from transformers import pipeline

device = "cuda" if torch.cuda.is_available() else "cpu"

llama_31 = "meta-llama/Meta-Llama-3.1-8B-Instruct"

generator = pipeline(model=llama_31, device=device, torch_dtype=torch.bfloat16)
````

````python
def generate_response(system_prompt, user_prompt):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ]

    generation = generator(
        messages,
        do_sample=False,
        temperature=1.0,
        top_p=1,
        max_new_tokens=200
    )

    return generation[0]['generated_text']
````

````python
import json

def chat_with_user(system_prompt, user_prompt):
    system_prompt = "You are a AI assistant who always responds in korean"
    ai_message = generate_response(system_prompt, user_prompt)
    return ai_message
````

````python
while True:
    user_prompt = input("USER > ")
    if user_prompt.lower() == "quit":
        break

    response = chat_with_user("", user_prompt)
    ai_message = json.dumps(response, ensure_ascii=False, indent=3)
    print(f" Llama3.1 > {ai_message} \n")
````
